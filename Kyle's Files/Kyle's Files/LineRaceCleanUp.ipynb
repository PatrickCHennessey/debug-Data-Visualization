```python
def binningFunc(col): # Granular bins for exponentially smaller, more popular songs. 
    col = float(col)
    if col > 50: # Top 100
        return 1
    elif col > 25: # Top 50
        return 2
    elif col > 10: # Top 25
        return 3
    elif col > 5: # Top 10
        return 4
    elif col > 1: # Top 5
        return 5
    else:
        return 6 # The Hottest of Hot Singles
```

```python
b_df5 = b_df5.assign(bins = lambda x: x["Weekly.rank"].apply(binningFunc))
```

```python
b_df5[b_df5["Weekly.rank"] == 26].head()
```

```python
b_df5[b_df5["Weekly.rank"] == 2].head()
```

```python
def popularityPoints(column):
    if column == 1:
        return (110) # Greatest Popularity metric for a week: 110
    else:
        return abs(100 - column) # Otherwise the inverse of rank. Larger Ranks -> Smaller popularity
```

```python
b_df5 = b_df5.assign(popPoints = lambda x: x["Weekly.rank"].apply(popularityPoints))
```

```python
b_df5["POP"] = b_df5["bins"] * b_df5["popPoints"] # Allow popularity to measured in a way other than just rankings alone. Exponentially leans toward higher binned songs.
# Gives another way to visualize. Best Weekly Score = 110 * 6 = 660
```

```python
def genreCounter(series, dictionary):
    for x in series:
        for g in x:
            if g not in dictionary.keys():
                dictionary[g] = 1
            else:
                dictionary[g] += 1
    return dictionary
```

```python
def genreMe(series): # SuperGenres
    col = series[-1]
    if "Reggae" in col:
        return 10
    elif "Electronic" in col:
        return 9
    elif "Folk" in col:
        return 8
    elif "R&;B" in col:
        return 7
    elif "Pop" in col:
        return 6
    elif "Country" in col:
        return 5
    elif "Rap" in col:
        return 4
    elif "Rock" in col:
        return 3
    elif "Pop-Rock" in col:
        return 2
    else:
        return 1
```

```python
dataframe_withGenres = dataframe2.assign(Master_Genre = lambda x: x.GenreArray.apply(genreMe))
```

```python
dataframe_withGenres.head()
```

```python
codeCarrier = {}
```

```python
def artistsCodes(series):
    codeList = []
    for x in series:
        if x in codeCarrier.keys():
            codeList.append(codeCarrier[x])
        else:
            codeCarrier[x] = hash(x)
            codeList.append(codeCarrier[x])
    return codeList
```

```python
dataframe_withGenres = dataframe_withGenres.assign(Artist_Code = lambda x: x.ArtistArray.apply(artistsCodes))
```
